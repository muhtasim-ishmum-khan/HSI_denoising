```python
# CONVERTING THE HSI IMAGES FROM HDR/RAW TO .MAT FORMAT
# imports::
import os 
import spectral
import numpy as np
from scipy.io import savemat
```

```python
# BATCH CONVERSION OF HSI TO .MAT FORMAT 

hdr_dir = "/Users/muhtasimishmumkhan/Desktop/499/hsi/data_custom/Day1/PlateA/capture"
mat_dir = "/Users/muhtasimishmumkhan/Desktop/499/hsi/data_custom/Weed_mat"

for file in os.listdir(hdr_dir):
    if file.endswith(".hdr"):
        hdr_path = os.path.join(hdr_dir, file)

        img = spectral.open_image(hdr_path)
        hsi_data = img.load().astype(np.float32)
        hsi_data = np.transpose(hsi_data, (2,0,1))

        save_name = os.path.splitext(file)[0] + ".mat"
        save_path = os.path.join(mat_dir, save_name)
        savemat(save_path, {"hsi": hsi_data})

        print(f"Saved: {save_name} ({hsi_data.shape})")
```
